 # Create Model

```python
# the first step divide your data
x=data.drop('Price',axis=1)
y=data['Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.80)
```

```python
model1=XGBRegressor()
```

```python
model1.fit(x_train,y_train)
```

```python
pre = model1.predict(x_test)
```

```python
metric1=r2_score(pre,y_test )#بدي افحص أدائه بعد ما تم تدريبه 
metric1
```

```python
metric2=mean_absolute_error(pre,y_test) # كم نسبة الايررور لهذا الموديل
metric2  # طبعا نسبة ممتازة جدا جدا 
```
